```python
# used for data visualization
import matplotlib.pyplot as plt
%matplotlib inline
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Weather Report')
x = [1, 2, 3, 4, 5, 6, 7] #day
y = [20, 35, 35, 40, 25, 15, 45] # temprature
# plot line graph
plt.plot(x, y, 'r*-')
```

```python
# axes, label, legend and grid

# data
week_days = [1, 2, 3, 4, 5, 6, 7]
max_temp = [20, 35, 25, 40, 25, 15, 45]
min_temp = [10, 30, 20, 35, 23, 12, 40]

# label for x-axis
plt.xlabel('Days')
# label for y-axis
plt.ylabel('Temperature')
# title of graph
plt.title('Weather Report')

# plotting 
plt.plot(week_days, max_temp, label = 'min')
plt.plot(week_days, min_temp, label = 'max')

# legend ( indicator and scaling )
plt.legend(loc='best', shadow=True, fontsize='large')

# grid ( for easy readability )
plt.grid()
```
