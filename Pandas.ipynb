```python
# for getting limited rows of data frames from BOTTOM
# by default it fetch 5 rows, using tail(n) method
nba_df.tail()
```

```python
# creating your own data file using dictonary
weather_data = {
    'date': ['1/10/2019','2/10/2019','3/10/2019','4/10/2019','5/10/2019','6/10/2019','7/10/2019','8/10/2019','9/10/2019'],
    'temperature': [32, 34, 35, 42, 1, 2, 33, 28, 3],
    'windspeed': [6, 8, 5, 9, 4, 8, 12, 10, 7],
    'event': ['Rain', 'Sunny', 'Rain', 'Sunny', 'Snow', 'Snow', 'Rain', 'Sunny', 'Snow']
}
# now converting it into data frame
weather_df = pd.DataFrame(weather_data)
weather_df
# getting data frames object in tabular form ( rows and columns)
```

```python
# diffrent way of creating data frame
#1. using csv file, pd.read_csv('path.csv')
#2. using excel file, pd.read_excel('path.xls')
#3. using python dictionary
#4. using tuple list
#5. using list of dictionary
```

```python
# 3. using python dictionary
weather_data = {
    'date': ['1/10/2019','2/10/2019','3/10/2019','4/10/2019'],
    'temperature': [35, 42, 1, 2],
    'windspeed': [6, 8, 5, 9],
    'event': ['Rain', 'Sunny', 'Rain', 'Sunny']
}
df1 = pd.DataFrame(weather_data)
df1
```

```python
# 4. using tuple list
weather_data = [
    ('1/10/2019',32,7,'rain'),
    ('1/10/2019',22,4,'snow'),
    ('1/10/2019',30,9,'sunny')
]
df2 = pd.DataFrame(weather_data, columns=["day", "temperature", "windspeed", "event"])
df2
```

```python
# 5. using list of dictionary
weather_data = [
    {'date': '1/10/2019', 'temperature': 22, 'windspeed': 6, 'event': 'Rain'},
    {'date': '2/10/2019', 'temperature': 35, 'windspeed': 4, 'event': 'Sunny'},
    {'date': '3/10/2019', 'temperature': 3, 'windspeed': 9, 'event': 'Snow'},
    {'date': '4/10/2019', 'temperature': 37, 'windspeed': 5, 'event': 'Sunny'}
]
df3 = pd.DataFrame(weather_data)
df3
```
